```python
# Read in data
data = pd.read_csv('emigrantpinoyeducrev2.csv', index_col=0, encoding='latin1')
```

```python
# Perform your data wrangling and descriptive stats to help
data.info()

data
```

```python
data_nototalspct = data.drop(columns=['TOTAL', '%'])


data_nototalspct
```

```python
data_nototalspct = data_nototalspct.drop('TOTAL', axis=0)
data_nototalspct
```

```python
df = data_nototalspct.transpose() 
df.info()
```

```python

plt.style.use('seaborn-colorblind')

df.plot(kind='bar', figsize=(25,15), stacked=True,  legend=['Attainment'])
```

```python
#above is too disaggregated to communicate well
```

```python
df
```

```python
df['No Response or Below School Age'] = df['Not of Schooling Age'] + df['Not Reported / No Response']
df['None or Informal Education'] = df['No Formal Education'] + df['Non-Formal Education ']
df['Up to Elementary'] = df['Elementary Level'] + df['Elementary Graduate']
df['Up to High School'] = df['High School Level'] + df['High School Graduate']
df['Up to Vocational'] = df['Vocational Level'] + df['Vocational Graduate']
df['Up to College'] = df['College Level'] + df['College Graduate']
df['Up to Post Graduate'] = df['Post Graduate Level'] + df['Post Graduate']
df

#this is to aggregate by <accomplishment
```

```python
df2 = df

df2.drop(columns=['Not of Schooling Age', 'No Formal Education','Elementary Level','Elementary Graduate','High School Level','High School Graduate','Vocational Level','Vocational Graduate','College Level','College Graduate','Post Graduate Level','Post Graduate','Non-Formal Education ','Not Reported / No Response'], inplace=True)

df2

#this is to drop the original source columns
```

```python

plt.style.use('Solarize_Light2')


df2.plot(kind='bar', figsize=(25,15), stacked=True,  legend=['Attainment'])

#initial bar plotting
```

```python
fig = plt.figure(figsize=(25, 15),dpi=300)

ax = df2.plot(kind='bar', stacked=True, figsize=(25, 15), color=['#009392','#39b185','#9ccb86','#e9e29c','#eeb479','#e88471','#cf597e'])

ax.set_title('Filipino Emigrant Educational Attainment', fontsize=20)
ax.set_ylabel('No. of Emigrant Filipinos', fontsize=15);
ax.set_xlabel('Years', fontsize=15);

#formatting
```
